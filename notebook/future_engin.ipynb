### Aggregate Credit and Debit transaction amount in the customer label and i did feature enginering the col i generate with decription describe below 
1. **Total Credit Amount**: Summed the values in the Credit column to determine the total credit transactions for each customer.
2. **Total Debit Amount**: Summed the values in the Debit column to determine the total debit transactions for each customer.
3. **Average Transaction Amount**: Calculated the mean of the Value column, which represents the average transaction value for each customer.
4.  **Transaction Count**: Counted the number of transactions per CustomerId using the TransactionId column.
5.  **Total Unique Account Count**: Counted the unique accounts associated with each customer using the AccountId column.
6.  **Standard Deviation of Transaction Amount**: Computed the standard deviation of the Amount column to assess the variability of transaction amounts. The ddof=0 parameter ensures a population standard deviation is calculated, and if there are fewer than two entries, NaN is replaced with zero. 
7. **Fraud Result**:This metric indicates the presence of fraud-related transactions. A value of 1 signifies that at least one transaction was flagged as fraudulent. 
8. **Channel-Specific Transaction Counts** : The sums for each specified channel (ChannelId_1, ChannelId_2, ChannelId_3, ChannelId_5) highlight the volume of transactions conducted through different channels. This can inform channel effectiveness and customer preferences. 
9. **Distinct Product Categories:** The count of unique product categories associated with transactions, which can reveal customer preferences and product diversity.
