#### The proxy label is assign using median treshoholds because it makes my data balanced
1. Threshold Definition: The function defines a threshold to classify customers into risk categories based on

5.  Implementation of Classification: The function utilizes NumPy's where method to assign the credit risk label ('Good' or 'Bad') based on the RFMS score.
